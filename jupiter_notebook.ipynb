```python
from PIL import Image, ImageOps
%pip install matplotlib
import matplotlib.pyplot as plt


```

```python
# Load an image
img_path = 'C:/Users/pinto/Desktop/Y3DIP/Lab 4 images/image1.jpg'  
img = Image.open(img_path)

# Display the image
plt.imshow(img)
plt.axis('off')  # Hide axis
plt.show()
```

```python
# Convert to grayscale
gray_img = ImageOps.grayscale(img)
plt.imshow(gray_img, cmap='gray')
plt.axis('off')
plt.show()
modified_img_path = 'C:/Users/pinto/Desktop/Y3DIP/image processing application images/new1.jpg'
gray_img.save(modified_img_path)

img = Image.open('your_image_path.jpg')  # Replace with your image path

# Convert to HSV
hsv_img = img.convert('HSV')

# Display the HSV image using Matplotlib
plt.imshow(hsv_img)
plt.axis('off')
plt.show()

# Save the converted image to the desired path
modified_img_path = 'C:/Users/pinto/Desktop/Y3DIP/image processing application images/new_hsv_image.jpg'
hsv_img.save(modified_img_path)

# Rotate the image
rotated_img = img.rotate(45)  # Rotate by 45 degrees
plt.imshow(rotated_img)
plt.axis('off')
plt.show()
modified_img_path = 'C:/Users/pinto/Desktop/Y3DIP/image processing application images/new.jpg'
rotated_img.save(modified_img_path)

# Crop the image
cropped_img = img.crop((100, 100, img.width - 100, img.height - 100))
plt.imshow(cropped_img)
plt.axis('off')
plt.show()

# Flip the image
flipped_img = ImageOps.flip(img)
plt.imshow(flipped_img)
plt.axis('off')
plt.show()
```
